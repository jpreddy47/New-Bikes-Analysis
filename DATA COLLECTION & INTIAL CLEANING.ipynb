## Sub Title : Data Collection & Initial Cleaning of Bikes Data

### Importing required libraries for data scraping from web

```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
```

#### Getting url

```python
url=requests.get("https://www.carandbike.com/new-bikes/models")
```

#### Response got from web is accessable if response is 200

```python
url
```

#### Beautifulsoup uses to extracting the text from the web

```python
soup=BeautifulSoup(url.text)
```

### Writing the codes for extracting the exact required data from the text 

```python
models=[]
# Scraping the data of bikes models from text
for i in soup.find_all("figcaption",class_="newmodel-bike__ttl"):
    a = i.text
    models.append(a.split("\n")[0])
    
len(models)


prices=[]
# Scraping the data of bikes prices from text
for i in soup.find_all("div",class_="newmodel-bike__ft grid grid__gap10"):
    a = i.text
    prices.append(a.split("\n")[3])
    
len(prices)

cc=[]
# Scraping the data of bikes CC from text
for i in soup.find_all("ul",class_="newmodel-bike__spec grid-flex grid-am h__mb15"):
    a = i.text
    cc.append(a.split("\n")[1])
    
len(cc)


mileage=[]
# Scraping the data of bikes mileage from text
for i in soup.find_all("ul",class_="newmodel-bike__spec grid-flex grid-am h__mb15"):
    a = i.text
    mileage.append(a.split("\n")[2])
    
len(mileage)


weight=[]
# Scraping the data of bikes weights from text
for i in soup.find_all("ul",class_="newmodel-bike__spec grid-flex grid-am h__mb15"):
    a = i.text
    weight.append(a.split("\n")[3])
    
len(weight)


rating=[]
# Scraping the data of bikes ratings from text
for i in soup.find_all("div",class_="newmodel-bike__rating h__mb15"):
    a = i.text
    rating.append(a.split("/")[0].split("(")[1])
    
len(rating)

reviews=[]
# Scraping the data of bikes reviews from text
for i in soup.find_all("span",class_="starrating__score"):
    a = i.text
    reviews.append(a.split("\n")[2])
    
reviews
```

### Converting all the data to a Dictionary & Creating a Dataframe by using the dictionary

```python
dict={"Model":models,
     "Prices":prices,
     "Bike CC":cc,
     "Mileage":mileage,
     "Weight":weight,
     "Ratings":rating,
     "Reviews":reviews}

df = pd.DataFrame(dict)
df
```

### url defined as x to get the all pages data 

```python
x="""https://www.carandbike.com/new-bikes/models/"""
```

### Writing the codes for extracting the data from the all pages of the urls

```python
models=[]
prices=[]
cc=[]
mileage=[]
weight=[]
rating=[]
reviews=[]

# code for printing all pages & getting access for text and loading data.
for i in range(2,32):
    xs = x+str(i)
    
    print(xs)
    page = requests.get(xs)
    soups = BeautifulSoup(page.text)
    
# Scraping the data of bikes models from text of all pages
    for i in soups.find_all("figcaption",class_="newmodel-bike__ttl"):
        a = i.text
        models.append(a.split("\n")[0])

# Scraping the data of bikes prices from text of all pages
    for i in soups.find_all("div",class_="newmodel-bike__ft grid grid__gap10"):
        a = i.text
        prices.append(a.split("\n")[3])
        
# Scraping the data of bikes CC from text of all pages
    for i in soups.find_all("ul",class_="newmodel-bike__spec grid-flex grid-am h__mb15"):
        a = i.text
        cc.append(a.split("\n")[1])

# Scraping the data of bikes mileage from text of all pages
    for i in soups.find_all("ul",class_="newmodel-bike__spec grid-flex grid-am h__mb15"):
        a = i.text
        mileage.append(a.split("\n")[2])

# Scraping the data of bikes weights from text of all pages
    for i in soups.find_all("ul",class_="newmodel-bike__spec grid-flex grid-am h__mb15"):
        a = i.text
        weight.append(a.split("\n")[3])

# Scraping the data of bikes ratings from text of all pages
    for i in soups.find_all("div",class_="newmodel-bike__rating h__mb15"):
        a = i.text
        rating.append(a.split("/")[0].split("(")[1])

# Scraping the data of bikes reviews from text of all pages
    for i in soups.find_all("span",class_="starrating__score"):
        a = i.text
        reviews.append(a.split("\n")[2])
    
len(reviews)
```

### Converting the data of all pages to another Dictionary & Creating another Dataframe by using the dictionary

```python
dict1={"Model":models,
     "Prices":prices,
     "Bike CC":cc,
     "Mileage":mileage,
     "Weight":weight,
     "Ratings":rating,
      "Reviews":reviews}

df1 = pd.DataFrame(dict1)
df1
```

### Joining the two Dataframes into one DataFrame using concat method

```python
final_df = pd.concat([df,df1],axis=0)
```

```python
final_df
```

#### Resetting the index after concating the dataframes

```python
final_df = final_df.reset_index()
```

```python
final_df
```

#### Dropping the duplicate index from dataframe

```python
final_df = final_df.drop('index',axis=1)
```

```python
final_df
```

```python
# removing the "₹" symbol in prices column
a = []
for i in final_df['Prices']:
    a.append(i.replace("₹",""))
final_df['Prices']  = a

# removing the "kg" in weight column
b = []
for i in final_df['Weight']:
    b.append(i[0:3])
final_df['Weight'] = b

# removing the "km/l" in Mileage column
c = []
for i in final_df['Mileage']:
    c.append(i.split()[0])
final_df['Mileage'] = c

# removing the "CC" in Bike CC column
d = []
for i in final_df['Bike CC']:
    d.append(i.split()[0])
final_df['Bike CC'] = d

# Creating a new column called "Brand" by using the existing "Model" column
e = []
for i in final_df['Model']:
    e.append(i.split()[0])
final_df['Brand'] = e

# removing the "ratings" in Reviews column
f = []
for i in final_df['Reviews']:
    f.append(i.split()[0])
final_df['Reviews'] = f
```

#### Output of the dataframe after cleaning

```python
final_df
```

### Saving the dataframe to a csv file 

```python
final_df.to_csv("totalcleaneddata.csv")
```
