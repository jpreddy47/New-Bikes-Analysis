## Sub Title : Data Visualization of Bikes Data

### Importing required libraries for data visualization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Loading the file

```python
df=pd.read_csv(r"C:\\Users\\Manisha Reddy\\DATA SCIENCE\\WEB SCRAPING\\Web Scraping PROJECT\\cleaneddata.csv")
```

```python
df
```

### Pairplot for co-relation of each columns

```python
sns.pairplot(df,hue="Brand")
```

### Highest price in over all brands

```python
plt.figure(figsize=(15,6))
sns.barplot(x="Brand",y="Prices",data=df,ci=None)

plt.xticks(rotation=90)
plt.show()
```

### Using box plot showing the reviews in each rating

```python
plt.figure(figsize=(15,6))
sns.boxplot(x="Ratings",y="Reviews",data=df)

plt.xticks(rotation=90)
plt.show()
```

### Most of the count of ratings

```python
plt.figure(figsize=(15,6))
sns.countplot(x="Ratings",data=df)

plt.xticks(rotation=90)
plt.show()
```

### Count of Mileage

```python
plt.figure(figsize=(18,6))
sns.countplot(x="Mileage",data=df)

plt.xticks(rotation=90)
plt.show()
```

### Count of Models in each Brand

```python
plt.figure(figsize=(15,6))
sns.countplot(x="Brand",data=df)

plt.xticks(rotation=90)
plt.show()
```

### Most reviewed Brand

```python
plt.figure(figsize=(15,6))
sns.barplot(x="Brand",y="Reviews",data=df,ci=None)

plt.xticks(rotation=90)
plt.show()
```

### High Prices Brand or Costliest Brand overall

```python
plt.figure(figsize=(15,6))
sns.barplot(x="Brand",y="Prices",data=df,ci=None)

plt.xticks(rotation=90)
plt.show()
```

### Highest CC Bike Brand

```python
plt.figure(figsize=(15,6))
sns.scatterplot(x="Brand",y="Bike CC",data=df)

plt.xticks(rotation=90)
plt.show()
```
