### Dropping the unnamed duplicate index column 

```python
df.drop(df.filter(regex="Unname"),axis=1, inplace=True)
```

### Checking the info to know the Datatype of the each column

```python
df.info()
```

#### Found that Prices & reviews columns are in Object, so we need to convert them to float or int...

#### Converting the Dtype of Prices from Object to Float by creating a function & apply...

```python
def splitfloat(a):
    return float("".join(a.strip().split(",")))
```

```python
df["Prices"]=df["Prices"].apply(splitfloat)
```

```python
df.head(10)
```

#### Converting the Dtype of Reviews from Object to int by creating a function & apply...

```python
def splitint(a):
    return int("".join(a.strip().split(",")))
```

```python
df["Reviews"]=df["Reviews"].apply(splitint)
```

```python
df.head(10)
```

#### Checking the info to know whether the Dtype is changed or not...

```python
df.info()
```

### Reindexing the columns for better understanding

```python
df = df.reindex(columns=['Brand','Model','Bike CC','Weight','Mileage','Prices','Ratings','Reviews'])
```

```python
df
```

### saving the file to csv to not to get duplicate index (by using index=false)

```python
df.to_csv('cleaneddata.csv', index=False)
```
